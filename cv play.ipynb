```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

remote = cv2.imread("remote.jpeg")
gremote = cv2.cvtColor(remote, cv2.COLOR_BGR2GRAY)
eremote = cv2.Canny(gremote, 50, 120)
minLineLength = 60
maxLineGap = 5
lines = cv2.HoughLinesP(eremote, 1, np.pi/180.0, 20,
                        minLineLength, maxLineGap)
for x1, y1, x2, y2 in lines[0]:
    cv2.line(remote, (x1, y1), (x2, y2), (255,0,0),2)
plt.imshow(eremote)

```
